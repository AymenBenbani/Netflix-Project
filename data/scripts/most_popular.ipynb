```python
import pandas as pd

# Charger le fichier csv
data = pd.read_csv('../raw_data/most_popular.csv')
data
```

```python
# Remplacer les valeurs non 'en' par 'no_en' dans la colonne 'original_language'
data.loc[data['original_language'] != 'en', 'original_language'] = 'Non-English'
data.loc[data['original_language'] == 'en', 'original_language'] = 'English'
data
```

```python
# Supprimer les colonnes non interessante
colonnes_a_supprimer = ['genre', 'overview', 'release_date', 'vote_count']
data = data.drop(colonnes_a_supprimer, axis=1)
data
```

```python
print(data.isna().sum())
print("There are {} missing values in the dataframe.".format(data.isna().sum().sum()))
```

```python
# How many duplicate values ?
print("There are {} duplicate values in the dataframe.".
     format(data.duplicated().sum()))
```

```python
# Enregistrer le DataFrame mis à jour dans un fichier CSV
data.to_csv("../suitable_data/clean_most_popular.csv", index=False)
```
